```python
def get_flow_run_info(flow_id,token):
  '''function to get the latest job_id run from the flow_id '''

  #define endpoit for dataprep API call
  url   = 'https://api.clouddataprep.com/v4/jobGroups?limit=1000,sort=-createdAt'
  
  print(f"Target flow id {flow_id}")

  #assemble authentication for api call
  head = {'Authorization': 'Bearer {}'.format(token)}

  req = urllib.request.Request(url,headers=head)

  #get data from api call
  with urllib.request.urlopen(req) as url:
    data = json.loads(url.read().decode())

  print(len(data['data']))

  #search for the flow_id in the output of the call
  for i in range(len(data['data'])):
    
    if data['data'][i]['wrangledDataset']['flow']['id'] == flow_id:
      print(f"index : {i}")
      print(f"job_id : {data['data'][i]['id']}")
      print(f"flow_name : {data['data'][i]['wrangledDataset']['flow']['name']}")
      print(f"flow created at : {data['data'][i]['createdAt']}")

      job_id    = data['data'][i]['id']
      flow_code = data['data'][i]['wrangledDataset']['flow']['name'].split('__')[0] 

      return job_id,flow_code
```
